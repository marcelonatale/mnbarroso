```python
inp1 = [1, 3, 6]
inp2 = [1, 2, 5, 14]
inp3 =[]
i, j = 0,0       # while loop command
while i < len(inp1):   
    while j < len(inp2):
        if i < j :
            inp3.append(inp1[i])
            i += 1
        else:
            inp3.append(inp2[j])
            j += 1
print(inp3)
```
