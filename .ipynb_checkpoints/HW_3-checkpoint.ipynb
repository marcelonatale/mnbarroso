Identify all the unique values in the dictionary  
inp: ```inp = {1:2, 2:3, 3:4, 4:2, 5:2}```  
out: ```[2 ,3 ,4]```

return a list of the values of a dictionary when its corresponding key is an integer less than 10.  
inp: ```{'Econ': '611', 1:2, 14:15, 6:'14'}```  
out: ```[2, 14]```

````python
fact_ = {'Econ': '611', 1:2, 14:15, 6:'14'}
geez_ = []

for k, v in fact_.items():
    if type(k) == int and k < 10:
         geez_.append(v)
print(geez_)


#I have tried this same code in different notebooks and worked, i couldn't find why it didn't in this one.
````

Given a string, find the word with the highest count.  
a. Separate the string into words  
b. keep a count of the words  
c. return the word with the highest count  
inp: ```"I went to school to study."```  
out: ```"to: 2"```

````python
from collections import Counter
inp = "I went to school to study."
words = inp.split()
print(words)
counts = Counter(words)
print(counts)
maximum = max(counts, key=counts.get)  # Could use min or max
print(maximum,counts[maximum])
````

#### 2. Create a new dictionary that is grouped at the host level. 
<span style="color:red">10/10</span>  
The output should be ```{host_id:[{Dictionary of listing 1}, {listing 2},...]}```   
How many listings do hosts have on average?

````python
host_list = []

for i in data:
    host_list.append(data[i]['host_id'])

host_dict = {}
for i in set(host_list):
    host_dict[i] = 0
for i in data:
    host = data[i]['host_id']
    host_dict[host] += 1
````

````python
total = 0
for i in host_dict:
    total += host_dict[i]

average = total/len(host_dict)

print(average)
````
