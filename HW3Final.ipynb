```python
#Questions 1
with open ('/Users/marcelonatale/Desktop/ECON611/ECON611/util_python/gen_hw_3.py'):
    data = load_data()
hoods_list = []

for i in data:
    hoods_list.append(data[i]['neighbourhood_group'])

hoods_dict = {}   
for i in set(hoods_list):
    hoods_dict[i] = 0
for i in data:
    hood = data[i]['neighbourhood_group']
    hoods_dict[hood] += 1
hoods_dict
```

```python
type_list = []

for i in data:
    type_list.append(data[i]['room_type'])

type_dict = {}   
for i in set(type_list):
    type_dict[i] = 0
for i in data:
    type = data[i]['room_type']
    type_dict[type] += 1
type_dict
```

```python
#2
host_list = []

for i in data:
    host_list.append(data[i]['host_id'])

host_dict = {}
for i in set(host_list):
    host_dict[i] = 0
for i in data:
    host = data[i]['host_id']
    host_dict[host] += 1
host_dict
print(host_dict)
total = 0
for i in host_dict:
    total += host_dict[i]

average = total/len(host_dict)
```
